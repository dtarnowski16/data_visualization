```python
#ISSUES
# 1. location of medians is off
# 2. Plot line on top of points instead of underneath


# Create a new figure of size 8x6 points, using 70 dots per inch
plt.figure(figsize=(8,4), dpi=75)

# Create a new subplot from a grid of 1x1
plt.subplot(111)

graph = sns.swarmplot(y='Comp_Index', x='Competition', data=df,size=5, color = 'black')
graph.set(ylabel='Competitive Index',xlabel='Competition Identifier',title='All replicates',ylim=[-3,3])

graph = sns.stripplot(x='Competition',y='Comp_Index',data=medians,color='r',size=6)
#graph = plt.scatter(x='Competition',y='Comp_Index',data=medians, color='red', marker='_', s=500, linewidths=2, alpha=1.0)

# Show result on screen
plt.show()
```

```python
df_trunc = pd.read_excel('data/2016_12_15_qrr1binK_AllData.xlsx')
df_trunc = df_trunc[df_trunc.Competition <= 4]
trunc_medians = df_trunc.groupby(['Competition'], as_index=False).median()

# Create a new figure of size 8x6 points, using 70 dots per inch
plt.figure(figsize=(8,4), dpi=75)

# Create a new subplot from a grid of 1x1
plt.subplot(111)

graph = sns.swarmplot(y='Comp_Index', x='Competition', data=df_trunc,size=5, color = 'black')

graph = sns.stripplot(x='Competition',y='Comp_Index',data=trunc_medians,color='r',size=6)
#graph = plt.scatter(x='Competition',y='Comp_Index',data=medians, color='red', marker='_', s=500, linewidths=2, alpha=1.0)

graph.set(ylabel='Competitive Index',xlabel='Competition Identifier',title='All replicates',ylim=[-3.5,3.5])

# Show result on screen
plt.show()
```

# Removing Individual experiments

```python
# No number 1
df_trunc = df[df.Experiment != 1]
trunc_medians = df_trunc.groupby(['Competition'], as_index=False).median()

# Create figure
plt.figure(figsize=(8,4), dpi=75)
plt.subplot(111)

graph = sns.swarmplot(y='Comp_Index', x='Competition', data=df_trunc,size=5, color = 'black')
graph = sns.stripplot(x='Competition',y='Comp_Index',data=trunc_medians,color='r',size=6)
graph.set(ylabel='Competitive Index',xlabel='Competition Identifier',title='No 1',ylim=[-3.5,3.5])

plt.show()
trunc_medians
```

```python
# No number 6
df_trunc = df[df.Experiment != 6]
trunc_medians = df_trunc.groupby(['Competition'], as_index=False).median()

# Create figure
plt.figure(figsize=(8,4), dpi=75)
plt.subplot(111)

graph = sns.swarmplot(y='Comp_Index', x='Competition', data=df_trunc,size=5, color = 'black')
graph = sns.stripplot(x='Competition',y='Comp_Index',data=trunc_medians,color='r',size=6)
graph.set(ylabel='Competitive Index',xlabel='Competition Identifier',title='No 6',ylim=[-3.5,3.5])

plt.show()
trunc_medians
```
